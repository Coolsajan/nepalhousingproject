### PRICE

```python
def pricing(price):
    price=price.lower()
    try:
        if "cr" in price:
            split_cr=price.split()
            price=float(split_cr[1])*10000000
            price=int(price)
    except :pass
    try:
        if "lac" in price:
            split_lac=price.split()
            price=float(split_lac[1])*1000000
            price=int(price)
    except :pass
    try:
        if "," in price:
            price=price.split()[1]
            price=price.split(",")
            price=int("".join(price))
    except:pass
    try:
        if "price on call"  in price:
            price=np.nan
    except: pass
    try:
        if "/aana"  in price:
            price=np.nan
    except: pass
    try:
        price=int(price.split()[1])
    except:pass
    
    return price

df['PRICE']=df['PRICE'].apply(pricing)
```

### LAND

```python
df['LAND AREA'].unique()
```

```python

anna_to_sqf=342.25
kattha_to_sqf=3645.013

def land(area):
    try:    
        area=area.lower()
        try:
            if "aana" in area:
                area=area.split()[0]
                area=float(area)*anna_to_sqf
        except:pass
        try:
            if "kattha" in area:
                area=area.split()[0]
                area=float(area)*kattha_to_sqf
        except: pass
        try:
            if "sq" in area:
                area=float(area.split()[0])
        except : pass
    except:pass
    return area


df['LAND AREA']=round(df['LAND AREA'].apply(land),2)
```

### ROAD ACESS

```python
df["ROAD ACCESS"].unique()
```

```python
meter_to_ft=29.5276
def road(road_size):
    try:
        road_size=road_size.lower()
        try:
            road_size=road_size.split()
            if len(road_size)==2:
                if '/' in road_size[0]:
                    road_size=road_size[0].split("/")[0]
                    road_size=int(road_size)
                elif "-" in road_size[0]:
                    road_size=road_size[0].split("-")[0]
                    road_size=int(road_size)
                elif "meter" in road_size:
                    road_size=float(road_size[0])*meter_to_ft
                    road_size=int(round(road_size))
                else :
                    road_size=int(road_size[0])
            else:
                return int(road_size[0])
        except:pass
    except:pass
    return road_size

df['ROAD ACCESS']=df['ROAD ACCESS'].apply(road)
```

### BUILD UP YEAR

```python
def build_year(year):
    try:
        year=year.split()[0]
        year=2081-int(year)
    except:
        year=None
    return year
df["BUILT YEAR"]=df["BUILT YEAR"].apply(build_year)
```

```python
def facing_structure(facing):
    try:
        facing=facing.lower()
        try:
             if '/' in facing:
                 facing=facing.split('/')
                 facing.sort()
                 facing="-".join(facing)
        except: pass

        try:
            if '-' in facing:
                facing=facing.split('-')
                facing.sort()
                facing='-'.join(facing)
        except: pass
        try:
            facing=facing.split()
            if len(facing)==2:
                facing="-".join(facing)
            else:pass
                
        except:pass
    except :
        None

    return facing

df['FACING']=df['FACING'].apply(facing_structure)
df['FACING']=df['FACING'].explode()
```
